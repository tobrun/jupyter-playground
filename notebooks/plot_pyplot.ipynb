# Get data

Download some public data from public APIs org. Parse data into a Pandas DataFrame.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

url = "https://api.publicapis.org/entries"
response = requests.get(url)
data = response.json()
api_list = data['entries']

df = pd.DataFrame(api_list)
```

# Plot data 

```python
# Create a bar plot of the number of APIs in each category
df['Category'].value_counts().plot(kind='bar')
plt.title('Number of APIs in each category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Create a pie chart of the proportion of APIs in each category
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of APIs in each category')
plt.show()
```

```python
# Create a boolean column for HTTPS
df['HTTPS'] = df['HTTPS'].apply(lambda x: True if x == 'True' else False)

# Create a boolean column for Link
df['Link'] = df['Link'].apply(lambda x: True if x else False)

# Create a line plot of the number of APIs over time
df.groupby(['API']).size().plot()
plt.title('Number of APIs over time')
plt.xlabel('API')
plt.ylabel('Count')
plt.show()
```
