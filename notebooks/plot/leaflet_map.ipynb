# Download and parse data
Get earthquake data and read into GeoPandas dataframe

```python
import geopandas as gpd

# Fetch the earthquake data parse as GeoDataFrame
gdf = gpd.read_file('https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson')
```

# Plot to Map
Use Python Folium library that gives access to Leaflet JavaScript library.

```python
import folium

map = folium.Map(location=[gdf['geometry'].y.mean(), gdf['geometry'].x.mean()], zoom_start=2)
for i in range(len(gdf)):
    mag = gdf.iloc[i]['mag']
    if mag < 4:
        fill_color = 'green'
    elif mag >= 4 and mag < 5:
        fill_color = 'orange'
    else:
        fill_color = 'red'
    folium.CircleMarker(location=[gdf.iloc[i]['geometry'].y, gdf.iloc[i]['geometry'].x], radius=gdf.iloc[i]['mag'], fill_color=fill_color, color=fill_color, fill_opacity=0.7).add_to(map)
map
```
